Когда происходит вызов метода класса из объекта экзмемпляра класса интерпретатор автоматически, в качестве первого аргумента передает объект экземпляра из которого он был вызван. Параметр self.

```python
class Point:
    color = 'red'
    circle = 2
    
    def set_coords(self):
        print(f'Вызов метода set_coords, экземпляр: {self}!')

pt = Point()
```

Если мы внутри класса хотим объявить какой-нибудь метод и вызывать его через объекты этого класса, то в качестве первого аргумента нужно обязательно указать параметр self, который ссылается на экземпляр из которого он был вызван

```python
pt.set_coords()
```

```python
pt
```

```python
Point.set_coords()
```

```python
Point.set_coords(pt) # вызывая из класса необходимо явно передать экземпляр
```

```python
class Point:
    color = 'red'
    circle = 2
    
    def set_coords(self, x, y):
        self.x = x
        self.y = y
        
pt = Point()
pt.set_coords(1, 2)
```

```python
pt.__dict__
```

Таким образом, мы можем работать с локальными атрибутами конкретного экземпляра класса. Иначе говоря, когда мы создаем несколько экземпляров одного и того же класса, то методы класса(например, set_coords) не копируются в отдельные экземпляры, а остается в классе, и уже благодаря параметру self мы знаем с каким экземпляром класса в данный момент работает метод класса. 

```python
class Point:
    color = 'red'
    circle = 2
    
    def set_coords(self, x, y):
        self.x = x
        self.y = y
        
    def get_coords(self):
        return (self.x, self.y)
    
pt1 = Point()
pt2 = Point()
pt1.set_coords(1, 2)
pt2.set_coords(10, 20)
```

```python
pt1.get_coords()
```

```python
pt2.get_coords()
```

Все имена методов также являются атрибутами класса.
