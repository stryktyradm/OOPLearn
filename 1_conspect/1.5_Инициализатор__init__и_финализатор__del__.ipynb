```python
class Point:
    color = 'red'
    circle = 2
    
    def __init__(self, x, y):
        print('__init__ call')
        self.x = x
        self.y = y
        
    def __del__(self):
        print(f'Удаление экземпляра {str(self)}')
    
    def set_coords(self, x, y):
        self.x = x
        self.y = y
        
    def get_coords(self):
        return (self.x, self.y)
    
pnt = Point(2, 3)
print(pnt.__dict__)
```

Когда и в какой момент происходит удаление объектов?

Сборщик мусора - алгоритм интерпретатора Python, который отслеживает объекты, которые как только становятся ненужными удаляются.

Пока на какой либо объект ведет хотя бы одна внешняя ссылка он считается нужным, но как только пропадает хотя бы одна внешняя ссылка, но как только все внешние ссылки пропадают, например:

pt = Point()
pt = 0

Объект pt(экземпляр класа Point) будет удален, а память занимаемая им освобождена.
