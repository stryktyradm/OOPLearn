@classmethod

```python
class Vector:
    MIN_COORD = 0 # атрибуты принадлежащие классу
    MAX_COORD = 100
    
    @classmethod # объявление метода класса
    def validate(cls, arg): # параметр cls является ссылкой на текущий класс(Vector)
        return cls.MIN_COORD <= arg <= cls.MAX_COORD
        
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def get_coord(self):
        return self.x, self.y
```

Метод класса работает, в данном случае, исключительно с атрибутами класса, но не может обращаться к локальным атрибутам класса.

```python
print(Vector.validate(5)) # Метод класса вызывается через сам класс, при этом не указывая сам параметр(cls) класса
```

@staticmethod

Статические методы не имею доступа ни к атрибутам класса, ни к атрибутам экземпляров. Создается независимая функция, которая объявлена внутри класса, как правило, такая функция звязана с тематикой самого класса.

```python
class Vector:
    MIN_COORD = 0 
    MAX_COORD = 100
    
    @classmethod 
    def validate(cls, arg): 
        return cls.MIN_COORD <= arg <= cls.MAX_COORD
        
    def __init__(self, x, y):
        self.x = self.y = 0
        if self.validate(x) and self.validate(y):
            self.x = x
            self.y = y
        
    def get_coord(self):
        return self.x, self.y
    
    @staticmethod
    def norm_2(x, y):
        return x*x + y*y 
```

Таким образом, мы имеем сервисную или вспомогательную функцию

```python
v = Vector(1, 2)
print(Vector.norm_2(5, 6))
```
