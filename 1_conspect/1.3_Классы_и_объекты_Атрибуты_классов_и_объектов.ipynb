<h1> Классы и объекты. Атрибуты классов и объектов <h1>

```python
class Point: # простейшее определение класса.
    pass
```

```python
class Point:
    color = 'red' # атрибуты/свойства класса.
    circle = 2
```

Фактически класс образует пространство имен. Пространство имен класса Point содержит две переменные.

```python
Point.color = 'black'
```

```python
Point.color, Point.circle
```

```python
Point.__dict__ # посмотреть все атрибуты класса.
```

```python
a = Point() # создание экземпляра.
```

```python
b = Point()
```

```python
b.__dict__
```

```python
type(a)
```

```python
isinstance(a, Point)
```
