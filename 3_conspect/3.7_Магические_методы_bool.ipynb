Функция bool всегда возвращает True для любых объектов пользовательского класса.

Мы можем переопределить поведение функции bool(), либо:

* __ len __ () - вызывается функцией bool(), если не определен магический метод bool();

* __ bool __ () - вызывается в приоритетном порядке функцией bool();

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __len__(self):
        print('__len__')
        return self.x * self.x + self.y * self.y
```

```python
p = Point(3, 4)
```

```python
bool(p)
```

Функция bool от любого числового значения отличного от 0 возвращает True.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __len__(self):
        print('__len__')
        return self.x * self.x + self.y * self.y
    
    def __bool__(self):
        print('__bool__')
        return self.x == self.y
```

```python
p = Point(1, 1)
```

```python
bool(p)
```

Функция __ bool __ , если определен, то будет вызываться в приоритете.

```python
p = Point(1, 10)
bool(p)
```

Магический метод __ bool __ должен возвращать только булевы значения.

Обычно, функция bool() вызывается неявно, например, в условных операторах.

```python
p = Point(1, 10)
```

```python
if p:
    print('Правда')
else:
    print('Ложь')
```

Следовательно, данный метод может пригодиться, где требуется описать собственные проверки истинности или ложности объектов.
