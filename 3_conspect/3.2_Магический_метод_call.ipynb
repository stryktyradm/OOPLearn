**__ call __**

```python
class Counter:
    def __init__(self):
        self.__counter = 0
```

```python
c = Counter()
```

Когда производит вызов класса Counter(), то автоматически вызывается магический метод __ call __

```python
c()
```

Однако, как видно, экземпляры класса, подобно функциям мы вызывать не можем: 'Counter' object is not callable

```python
class Counter:
    def __init__(self):
        self.__counter = 0
        
    def __call__(self, *args, **kwargs):
        print('__call__')
        self.__counter +=1
        return self.__counter
```

```python
c = Counter()
```

```python
c()
```

Благодаря добавлению метода __ call __ экземпляры класса получают возможность быть вызванными подобно функциям. Классы, которые обладают данным поведением называются функторами.

```python
res = c()
```

```python
res
```

```python
class StripChars:
    def __init__(self, chars):
        self.__chars = chars

    def __call__(self, *args, **kwargs):
        if not isinstance(args[0], str):
            raise TypeError('Аргумент должен быть строкой')
        return args[0].strip(self.__chars)
```

```python
s = StripChars('! %#$:;.,')
```

```python
res = s('Hello world! ')
```

```python
res
```

```python
import math

class Derivate:
    def __init__(self, func):
        self.__fn = func
        
    def __call__(self, x, dx=0.0001, *args, **kwargs):
        return (self.__fn(x + dx) - self.__fn(x)) / dx
    
@Derivate    
def df_sin(x):
    return math.sin(x)

# df_sin = Derivate(df_sin)
df_sin(math.pi/3)
```
