Магические методы сравнения
* __ eq __ - ==
* __ ne __ - !=
* __ lt __ - <
* __ le __ - <=
* __ gt __ - >
* __ ge __ - >=


```python
class Clock:
    __DAY = 86400
    
    def __init__(self, seconds: int):
        if not isinstance(seconds, int):
            raise TypeError('Секунды должны быть целым числом')
        self.seconds = seconds % self.__DAY
```

```python
c1 = Clock(1000)
c2 = Clock(1000)
```

```python
c1 == c2  # оператор сработал и выдает значение False, так как сравниваются id объектов или адреса в памяти, но не время, которое они хранят
```

```python
class Clock:
    __DAY = 86400
    
    def __init__(self, seconds: int):
        if not isinstance(seconds, int):
            raise TypeError('Секунды должны быть целым числом')
        self.seconds = seconds % self.__DAY
        
    def __eq__(self, other):
        if not isinstance(other, (int, Clock)):
            raise TypeError('Операнд справа должен иметь тип int или Clock')
        sec = other if isinstance(other, int) else other.seconds
        return self.seconds == sec
```

```python
c1 = Clock(1000)
c2 = Clock(2000)
```

```python
c1 == c2
```

```python
c1 = Clock(1000)
c2 = Clock(1000)
```

```python
c1 == c2
```

```python
c1 == 1000
```

В свою очередь, оператор !=, если не определен явно будет сначала пытаться выполнить ==(__ eq __ ) и если же такой метод определен, то будет вызван not (obj1 == obj2), другими словами, инвертрует __ eq __

```python
class Clock:
    __DAY = 86400
    
    def __init__(self, seconds: int):
        if not isinstance(seconds, int):
            raise TypeError('Секунды должны быть целым числом')
        self.seconds = seconds % self.__DAY
        
    @classmethod
    def __verify_data(cls, other):
        if not isinstance(other, (int, cls)):
            raise TypeError('Операнд справа должен иметь тип int или Clock')
        return other if isinstance(other, int) else other.seconds
        
    def __eq__(self, other):
        sec = self.__verify_data(other)
        return self.seconds == sec
    
    def __lt__(self, other):
        sec = self.__verify_data(other)
        return self.seconds < sec
```

```python
c1 = Clock(1000)
c2 = Clock(2000)
```

```python
c1 < c2
```

```python
c1 > c2  # c2 < c1, если отсутсвует >
```
