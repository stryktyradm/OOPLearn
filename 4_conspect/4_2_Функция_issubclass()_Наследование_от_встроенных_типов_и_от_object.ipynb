По умолчанию пользовательский класс наследуется от базового класса object

```python
class Geom:
    pass
```

```python
Geom.__name__
```

Атрибуты, которые присущи классу Geom изначально наследуются или приобретаются от базового класса object

Объявление какого-либо класса эквивалентно объявлению:

```python
class Geom(object):
    pass
```

Таким образом, мы обеспичаваем стандартный базовый функционал для работы с классами.

```python
g = Geom()
g
```

Данная строчка появляется благодаря определению в классе object метода __ str __

```python
class Line(Geom):
    pass
```



*цепочка наследования будет выглядеть следующим образом*

Очевидно, что объекты класса Line будут иметь доступ ко всем атрибутам класса object

```python
l = Line()
print(l.__class__)
```

Мы можем определять является ли тот или иной класс подклассом другого класса, что делается с помощью специальной функции issubclass().

```python
issubclass(Line, Geom)
```

В качестве первого парметра указывается дочерний класс, а в качестве второго базовый. Данная функция работает только с классами, а не с объектами/экземплярами.

```python
issubclass(l, Geom)
```

Если же нам нужно проверить принадлежность объектов к тому или иному базовому классу, то следует использовать функцию isinstance()

```python
isinstance(l, Geom)
```

```python
isinstance(l, object)
```

```python
isinstance(Geom, object)
```

Все стандартные типы данных в Python также являются классами.

```python
issubclass(int, object)
```

Следовательно, ничто нам не мешает расишрять стандартные типы данных в Python(так как являются классами).

```python
class Vector(list):
    def __str__(self):
        return " ".join(map(str, self))
```

```python
v = Vector([1, 2, 3])
```
