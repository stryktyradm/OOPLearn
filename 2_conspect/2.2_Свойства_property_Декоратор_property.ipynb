Пользователю возможно придется иметь ввиду достаточно большое количество сеттеров и геттеров.

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
        
    def get_old(self):
        return self.__age
    
    def set_old(self, age):
        self.__age = age
        
    old = property(get_old, set_old) # считываем и записываем через единственный атрибут
```

```python
p = Person('Sergey', 20)
print(p.old)
p.old = 35 # почему происходит обращение к сеттеру, а не создается новое свойство? приоритет выше, чем у локального свойства?
print(p.old)
```

У объекта класса property имеются методы setter, getter, deleter, который представляют собой декотораторы и которые мы можем использоваться для работы с атрибутами или свойствами экземпляров класса.

<h2> Декораторы </h2>

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
        
    def get_old(self):
        return self.__age
    
    def set_old(self, age):
        self.__age = age
        
    old = property()
    old = old.setter(set_old)
    old = old.getter(get_old)
```

```python
p = Person('Sergey', 20)
p.old
p.old = 35
p.old
```

```python
class Person:
    def __init__(self, name, old):
        self.__name = name
        self.__old = old
     
    @property
    def old(self):
        return self.__old
    
    @old.setter
    def old(self, old):
        self.__old = old
```

```python
p = Person('Sergey', 35)
p.old = 35
p.old, p.__dict__
```
